```python
"""
**사용전 주의 사항

1,환경 설정값은 변경하시면 안됩니다,
2,이 파일의 실행은 함수 단위로 결과를 보여주는 식으로 되어있습니다 하나의 결과가 출력되면 esc를 누르고 다음으로 넘어가시면 됩니다
3.hsv로 보낼 구역 선택할때 최대한 몸 중앙과 조명이 없는 부분을 선택해 주세요

"""

import cv2
import numpy as np
import matplotlib.pyplot as plt

videoFile_path = 'D:/porject_for_jupyter/test/Term_Project/example2.mp4'

#캡처 감도 설정
thresh = 75 #픽셀 차이
max_diff = 2000 #차이가 나는 픽셀 갯수

#캡처 준비
cap_width = 404
cap_height = 720
frame_num_size = 150 # 인식후 읽어 들일 frame 갯수

one_frame =[None for i in range(frame_num_size)]#캡처한 프레임을 저장하는 리스트
diff_max_loc = [0 for i in range(10)]#변화 갯수가 제일 많은 frame 넘버를 저장
body_map = [[0 for col in range(cap_width)] for row in range(cap_height)]#인식한 몸 사진을 위치에 맞게 배열에 0,1로 저장 

#body_map에서 몸의 포인트를 찾기위한 준비
frequency = 5#검사하는 행의 빈도
body_center = 0
body_total_length = 168
shoulder_length = 0
upper_body_length = 0


def capture_body_image(videoFile):
    
    one_frame_gray =[None for i in range(frame_num_size)]
    diff = [None for i in range(frame_num_size)]
    diff_cnt = [0 for i in range(frame_num_size)]
    diff_max_loc = []
    
    cap = cv2.VideoCapture(videoFile)
    cap.set(cv2.CAP_PROP_FRAME_WIDTH,cap_width)#캡처 사진 크기 설정
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT,cap_height)

    if cap.isOpened():
        ret,one_frame[0] = cap.read()
    
        while ret:
            ret,one_frame[1] = cap.read()
            
            if not ret:
                break
            
            #정확한 판별을 위해 gray로 배경 설정,2개의 frame으로 변화 감지
            one_frame_gray[0] = cv2.cvtColor(one_frame[0],cv2.COLOR_BGR2GRAY)
            one_frame_gray[1] = cv2.cvtColor(one_frame[1],cv2.COLOR_BGR2GRAY)
        
            #변화 감지를 위한 두개의 프레임을 확인
            diff1 = cv2.absdiff(one_frame_gray[0],one_frame_gray[1])
        
            ret,diff1_t = cv2.threshold(diff1,thresh,255,cv2.THRESH_BINARY)
            diff[1] = diff1_t
        
            k = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
            diff[1] =cv2.morphologyEx(diff[1],cv2.MORPH_OPEN,k)
        
            diff_cnt[1] = cv2.countNonZero(diff[1])
        
            if diff_cnt[1] > max_diff:# 영상의 변화 감지
                #원본 배경화면과 비교해서 제일 cnt가 높은 frame 찾기
                #차이값을 저장
                max_cnt = diff_cnt[1]
                max_loc = 1;
                frame_count = 0
            
                for i in range(2,frame_num_size):
                    ret,one_frame[i] = cap.read()
                    if ret == 0:
                        break
                    one_frame_gray[i] = cv2.cvtColor(one_frame[i],cv2.COLOR_BGR2GRAY)
                
                    diff1 = cv2.absdiff(one_frame_gray[0],one_frame_gray[i])
                    ret,diff1_t = cv2.threshold(diff1,thresh,255,cv2.THRESH_BINARY)
                    diff[i] = diff1_t
        
                    k = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
                    diff[i] =cv2.morphologyEx(diff[i],cv2.MORPH_OPEN,k)
                
                    diff_cnt[i] = cv2.countNonZero(diff[i])
                    
                    #제일 높은 차이 뽑아내기
                    if max_cnt > diff_cnt[i]:
                        continue
                    else: 
                        max_cnt = diff_cnt[i]
                        max_loc = i
                    
                    #10개중 제일 큰거 저장 -> 리스트
                    if i%12 == 0:
                        diff_max_loc.append(max_loc)
                        max_cnt = 0
                    
                nzero = np.nonzero(diff[max_loc])# 차이가 있는 프레임 위치뽑아내기
                copy_img = one_frame[max_loc]#제일 차이가 많이나는 프레임 복사
                ret = 0#캡처 끝내기
                cv2.imshow("body_found",one_frame[max_loc])
        
        if cv2.waitKey(0)&0xFF == 27:
            cap.release()
            cv2.destroyAllWindows()
    
    return max_loc
```

```python
def make_body_map(max_loc): #원본 사진에서 몸 사진 분리 -> 역투영 방법 사용
            
    copy_img = one_frame[max_loc]#제일 차이가 많이나는 프레임 복사
    hsv_img = cv2.cvtColor(copy_img,cv2.COLOR_BGR2HSV)
    
    #사용자가 상체의 일정부분 선택
    (x,y,w,h) = cv2.selectROI('sample',copy_img,False)
    if x > 0 and y > 0: 
        if w >0 and h> 0:
            roi = copy_img[y:y+h,x:x+w]
            cv2.rectangle(copy_img,(x,y),(x+w,y+h),(0,0,255),1)
            hsv_roi = cv2.cvtColor(roi,cv2.COLOR_BGR2HSV)
            
            hist_roi = cv2.calcHist([hsv_roi],[0,1],None,[256 ,256],[0,256,0,256])#1채널로 우선 해보기 정확도 상승 -> 2채널 
            #hist_roi = cv2.calcHist([hsv_roi],[0],None,[256],[0,256])
    else:
        print("영역 선택이 올바르지 않습니다")
        return 0;
    #역투영 방법 사용
    bp = cv2.calcBackProject([hsv_img],[0,1],hist_roi,[0,256,0,256],1)
    #bp = cv2.calcBackProject([hsv_img],[0],hist_roi,[0,256],1)
    
    bp_sample = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
    cv2.filter2D(bp,-1,bp_sample,bp)
    _,mask = cv2.threshold(bp,1,255,cv2.THRESH_BINARY)
    mask
    final = cv2.bitwise_and(one_frame[max_loc],one_frame[max_loc],mask=mask)
    cv2.imshow('decomposit_img',final)
    
    #최종으로 뽑아낸 사진 final을 바탕으로 맵을 작성(검은색->0 나머지색 -> 1 인 2차원 배열)
    nzero = np.nonzero(mask)
    
    for i in range(len(nzero[0])):
        x = nzero[1][i]
        y = nzero[0][i]
        body_map[x][y] =1
        
    if cv2.waitKey(0)&0xFF == 27:
        cv2.destroyAllWindows()

```

```python
def find_body_point(max_loc):
    point = [[0 for col in range(6)] for row in range(int((cap_height/5))+2)]

    for i in range(0,cap_height,frequency):
        count_left = 0
        count_right = 0
        continue_num_left = 0
        continue_num_right = 0
        label_point = int(i/frequency)
        for j in range(cap_width):
            if int(cap_width/2) - 25 == j:#없는 픽셀로 취급
                break;
            
            if body_map[i][j] == 1:#왼쪽에서 오른쪽
                if continue_num_left == 1:
                    continue
                else:
                    point[label_point][0] = i
                    point[label_point][1] = j
                    continue_num_left = 1
        
            if body_map[i][cap_width - j-1] == 1:#오른쪽에서 왼쪽
                if continue_num_right == 1:
                    continue
                else :
                    point[label_point][2] = i
                    point[label_point][3] = cap_width - j-1
                    continue_num_right = 1
            
            if (continue_num_left + continue_num_right) == 2:#둘다 끝났을때
                break;
            
            else:
                if continue_num_left == 1:#왼쪽만 끝났을때
                    count_right +=1
                
                elif continue_num_right == 1:#오른쪽만 끝났을때
                    count_left +=1
            
                else:#둘다 안끝났을때
                    count_right +=1    
                    count_left +=1
        
        point[label_point][4] = count_left
        point[label_point][5] = count_right
    """
    point[][0] -> 왼쪽 x [1] -> 왼쪽 y [2] -> 오른쪽 x [3] -> 오른쪽 y [4] -> 왼쪽 길이 [5] -> 오른쪽 길이
    길이는 얼마나 갔는지(1을 만나기 전까지 얼마나 갔는지)를 나타내는 값
    """
    #어깨 라인을 추출하기 위해 count가 변동이 큰 지점을 찾아낸다
    min_length_left = (cap_width/2) -25
    min_length_right = (cap_width/2) +25
    final_point = [[0 for col in range(8)] for row in range(10)]
    cnt =0

    for i in range(len(point)):#왼쪽
        cal = min_length_left - point[i][4]
        if cal > 0:
            min_length_left = point[i][4]
            if cal > 20:
                for j in range(len(point[i])):
                    final_point[cnt][j] = point[i][j]
                final_point[cnt][6] = cal
                cnt +=1
            
    for i in range(len(point)):#오른쪽
        cal = min_length_right - point[i][5]
        if cal > 0:
            min_length_right = point[i][5]
            if cal > 20:
                for j in range(len(point[i])):
                    final_point[cnt][j] = point[i][j]
                final_point[cnt][7] = cal
                cnt +=1

    max_cal = 0
    max_cal_position = 20
   
    for i in range(len(final_point)):#오른쪽
        if max_cal < final_point[i][7]:
            if final_point[i][0] == 0 or final_point[i][2] == 0:
                continue
            else:
                max_cal  = final_point[i][7]
                max_cal_position = i
    
    max_cal = 0
    second_max_cal_position = 20
    
    for i in range(len(final_point)):#왼쪽
        if max_cal < final_point[i][6]:
            if final_point[i][0] == 0 or final_point[i][2] == 0:
                continue
            else:
                max_cal  = final_point[i][6]
                second_max_cal_position = i
                
    if max_cal_position == 20:
        max_cal_position = second_max_cal_position
        
    if second_max_cal_position == 20:
        second_max_cal_position = max_cal_position
                
    for i in range(len(final_point)):
        print(i,"번째 왼쪽 x,y:",final_point[i][0],",",final_point[i][1],",count:",final_point[i][4]," / 오른쪽 x,y:",final_point[i][2],",",final_point[i][3],",count:",final_point[i][5],final_point[i][6],final_point[i][7])
    
    #몸의 중앙 좌표 찾기
    body_center = int((final_point[max_cal_position][3]-final_point[second_max_cal_position][1])/2)+final_point[max_cal_position][1]
    
    #비율 계산을 통한 길이 측정
    body_length_rate = body_total_length / cap_height #비율
    global shoulder_length
    shoulder_length = int(body_length_rate * (abs(final_point[max_cal_position][3] - final_point[second_max_cal_position][1]))) + 1#어깨 길이
    
    cnt =0
    str_cnt =0
    upper_body_up = 0# 상체 제일 위 x 좌표
    upper_body_down = 0# 상체 제일 밑 x 좌표
    
    #상체의 길이를 구하는 코드
    for i in range(len(body_map)):
        if body_map[i][body_center] == 1:
            if str_cnt == 0:#시작
                upper_body_up = i
            str_cnt +=1
            if cnt != 0:
                cnt = 0
        else:
            if str_cnt == 0:# 시작 안함 
                continue
            else:#중간 노이즈 or 끝
                cnt+=1
                if cnt > 50:#끝
                    upper_body_down = i -50
                    break
    
    global upper_body_length
    upper_body_length = int(body_length_rate * (abs(upper_body_down - upper_body_up))) + 1
    
    #결과 출력
    #어깨
    cv2.circle(one_frame[max_loc],(final_point[second_max_cal_position][0],final_point[second_max_cal_position][1]), 10,(0,0,255), 3)
    cv2.circle(one_frame[max_loc],(final_point[max_cal_position][2],final_point[max_cal_position][3]), 10,(0,0,255), 3)
    #어깨
    #cv2.circle(one_frame[max_loc],(final_point[second_max_cal_position][0],final_point[second_max_cal_position][1]), 10,(0,0,255), 3)
    #cv2.circle(one_frame[max_loc],(final_point[second_max_cal_position][2],final_point[second_max_cal_position][3]), 10,(0,0,255), 3)
    #상체
    cv2.circle(one_frame[max_loc],(upper_body_up,body_center), 10,(0,0,255), 3)
    cv2.circle(one_frame[max_loc],(upper_body_down,body_center), 10,(0,0,255), 3)
    #각 포인트를 잇는 직선
    cv2.line(one_frame[max_loc],(upper_body_down,body_center),(final_point[second_max_cal_position][0],final_point[second_max_cal_position][1]),(255,0,0), 2)
    cv2.line(one_frame[max_loc],(upper_body_down,body_center),(final_point[max_cal_position][2],final_point[max_cal_position][3]),(255,0,0), 2)
    cv2.line(one_frame[max_loc],(0,body_center),(cap_height-1,body_center),(0,255,0), 2)
    cv2.line(one_frame[max_loc], (final_point[second_max_cal_position][0],final_point[second_max_cal_position][1]),(final_point[max_cal_position][2],final_point[max_cal_position][3]),(255,0,0), 2)
    cv2.line(one_frame[max_loc],(upper_body_down,body_center),(upper_body_up,body_center),(0,170,255), 2)
    
    cv2.imshow("show_check_point",one_frame[max_loc])
    
    if cv2.waitKey(0)&0xFF == 27:
        cv2.destroyAllWindows()
```

```python
max_loc = capture_body_image(videoFile_path)
make_body_map(max_loc)
find_body_point(max_loc)
```

```python
print("실제 측정값 -> 총장:",body_total_length,"cm , 어깨길이:",44,"cm , 상체길이:",48 ,"cm")
print("프로그램 연산 결과 -> 총장:",body_total_length,"cm , 어깨길이:",shoulder_length,"cm 오차 :",abs(shoulder_length - 44),", 상체길이:",upper_body_length,"cm 오차 :",abs(upper_body_length-48))
```

```python
def check_profit_size():
    
    size_table = [[0 for i in range(3)] for i in range(3)]
    
    size_table[0][0] = 'M' 
    size_table[0][1] = 71 
    size_table[0][2] = 48
    
    size_table[1][0] = 'L' 
    size_table[0][1] = 73 
    size_table[0][2] = 51 
    
    size_table[2][0] = 'XL' 
    size_table[0][1] = 75 
    size_table[0][2] = 54
    
    min_diff = 10
    loc_s = 0
    
    for i in range(len(size_table)):
        diff = size_table[i][2] - shoulder_length
        if diff > 0 and min_diff > diff:
            min_diff = diff
            loc_s = i
            break
    
    min_diff = 20
    loc_u = 0
    for i in range(len(size_table)):
        diff = size_table[i][1] - upper_body_length
        if diff > 0 and min_diff > diff:
            min_diff = diff
            loc_u = i
            break
    
    if loc_s == loc_u:
        return size_table[loc_s][0]
    else: 
        return 0
            
```

```python
print("사이즈 함수에 전달 -> 총장:",body_total_length,"cm , 어깨길이:",shoulder_length,"cm , 상체길이:",upper_body_length,"cm")
print("사이즈표를 참고한 추천 사이즈:",check_profit_size())
```
